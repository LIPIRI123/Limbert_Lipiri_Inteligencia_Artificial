```python
print("el numeros de horas dedicadas al estudio de una asignatura y la calificacion\n obtenida en el examen correspondiente de cinco personas es:")
import numpy as np
import matplotlib.pyplot as plt
hora=[20,16,34,10,23]
calificacion=[6.5,6,8,4,7]

n=len(hora)
n1=len(calificacion)
hora=np.array(hora)
calificacion=np.array(calificacion)
#sumando los vectores de x y y
sumhora=sum(hora)
sum_calificacion=sum(calificacion)

#obteniendo los promedios
promhora=sumhora/n
prom_calificacion=sum_calificacion/n1

#elevando al cuadrado la parte de abajo
parte_abajo=sum(pow(hora-promhora,2))

#para formula uno
#suCu=sum(pow(hora,2))
#sumaxy=sum(hora*calificacion)
#b=(sumhora*sum_calificacion-n*sumaxy)/(sumhora*sumhora-n*suCu)


#para formula dos
parte_abajo=sum(pow(hora-promhora,2))
parte_arriba=sum((hora-promhora)*(calificacion-prom_calificacion))
b=(parte_arriba)/(parte_abajo)

#print("b = ",b)
a=prom_calificacion-b*promhora
#print("a = ",a)
print("a) Diagrama de dispersion")
plt.plot(hora,calificacion,'o',Label='datos')
#plt.plot(hora,a+b*hora,Label='Ajuste')
plt.xlabel('horas de estudio')
plt.ylabel('calificacion obtenida')
plt.grid()#para poner al grafico lineas
plt.legend()
plt.show()
print("b) Calcular la recta de regresion de y sobre x")
plt.plot(hora,calificacion,'o',Label='datos')
plt.plot(hora,a+b*hora,Label='el mejorAjuste')
plt.xlabel('horas de estudio')
plt.ylabel('calificacion obtenida')
plt.grid()#para poner al grafico lineas
plt.legend()
plt.show()

print("c) calcular la calificacion estimada para una persona que hubiese estudiadado 28 horas")
operacion=a+b*28
operacio=a+b*34
print("                 la calificacion es: ",operacion)
print("d) comentar y concluir")
print("     la calificacion al estudio de 34 horas para el estudiante sera de: ",operacio)

print("     la calificacion al estudio de 28 horas para el estudiante sera de: ",operacion)
```
